```python
import json
import requests
import pandas as pd
```

```python
with open('apikey.txt', 'r') as f:
    API_KEY=f.read()
URL_BASE='https://www.googleapis.com/youtube/v3/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
```

```python
def getData(api, *args):
    URL_FULL=URL_BASE+api+'?'+'&'.join(args)+'&key='+API_KEY
    return(URL_FULL)
```

```python
def get_comments(searchterm, nVideos, nComments, searchtermComments=''):
    url=getData('search', 'part=snippet','q='+searchterm,'order=relevance','relevanceLanguage=en','maxResults='+str(nVideos))
    response=requests.get(url, headers=headers)
    data=json.loads(response.text)
    search_table=pd.DataFrame({'videoId':[i['id']['videoId'] for i in data['items']],'title':[i['snippet']['title'] for i in data['items']]})

    
    comments_table=pd.DataFrame()
    for i in search_table.videoId:
        url=getData('commentThreads', 'part=snippet,id','videoId='+i,'order=relevance','searchTerms='+searchtermComments,'maxResults='+str(nComments))
        response=requests.get(url, headers=headers)
        data=json.loads(response.text)
        if 'items' in data:
            comments_dict={'textOriginal':[i['snippet']['topLevelComment']['snippet']['textOriginal'] for i in data['items']]}
            comments_table=comments_table.append(pd.DataFrame(comments_dict))
    comments_table['source']='YouTube'
    return(comments_table.reset_index(drop=True))
```
